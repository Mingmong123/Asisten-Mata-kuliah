#### Distribusi Nilai
- 75 persen berprestasi
- 25 persen kesulitan



```python
#proof of concept utk data lama
for i in range(0,len(df)):
    
    temp_avg = 0
    #implementasi mahasiswa berprestasi atau kesulitan
    nilai_min = 2 if i%4==0 else 5
    nilai_maks = 10 if i%4==0 else 12
    
    
    #print(nilai_min)
    df.iloc[i,0]=i #no. mahasiswa

    for j in range(1,len(df.columns)-5,5):
        
        tp1 = randint(nilai_min,nilai_maks)
        tp2 = randint(max(1,tp1-4),min(nilai_maks,tp1+4))
        uts = randint(max(1,min(tp1,tp2)-4),min(nilai_maks,min(tp1,tp2)+5))
        uas = randint(max(1,min(tp1,tp2,uts)-4),min(nilai_maks,min(tp1,tp2,uts)+5))

        rata2 = round((tp1+tp2+uts+uas)/4)

        df.iloc[i,j] = tp1
        df.iloc[i,j+1] = tp2 
        df.iloc[i,j+2] = uts
        df.iloc[i,j+3] = uas
        df.iloc[i,j+4] = rata2
        
        #print(tp1, tp2, uts, uas, rata2)

        
```

```python
#Topik Skripsi di-random dari 3 top pilihan
rata2 = ['No Mhs']
for i in df.columns:
    
    if(str(i).lower().find('rata-rata')!=-1):
        rata2.append(i)

```

```python
df_rata2 = df[rata2]
df_rata2
```

```python
#memilih rekomendasi skripsi - data dummy
#random sort 3 nilai terbaik, ambil mata kuliahnya
for index,row in df_rata2.iterrows():
    row_mhs = dict(row)
    no_mhs = row_mhs['No Mhs']
    del row_mhs['No Mhs']
    
    row_mhs = sorted(row_mhs.items(), key= lambda i:i[1], reverse=True) #Thanks FreeCodeCamp https://www.freecodecamp.org/news/sort-dictionary-by-value-in-python/
    row_mhs = dict(row_mhs)
    
    top_3_nilai = list(row_mhs.keys())[:3]
    topik_skripsi = str(top_3_nilai[randint(0,2)])
    df.loc[index, 'Topik Skripsi'] = topik_skripsi.replace(' Rata-Rata','')
    
df

```
